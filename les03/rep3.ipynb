```python
import numpy as np

def create_matrix_and_find_extremes(size=(10, 10), low=1, high=101):
    matrix = np.random.randint(low, high, size=size)
    return matrix, matrix.min(), matrix.max()

matrix, min_val, max_val = create_matrix_and_find_extremes()

print("Массив:\n", matrix)
print(f"Минимум: {min_val}")
print(f"Максимум: {max_val}")
```

```python
import numpy as np

z = np.random.random((10,2))
polar = np.zeros((10, 2))

for i in range(10):
    polar[i][0] = np.sqrt(z[i][0]**2 + z[i][1]**2)
    polar[i][1] = np.arctan2(z[i][1], z[i][0])

print(polar)
```

1. Какой dtype получат массивы?
np.array([1, 4, 2, 5, 3])                     # int64
np.array([1., 4, 2, 5, 3])                    # float64
np.array([1, 4, 2, 5, 3], dtype='float128')   # float128
np.array([1., 4, 2, 5, 3], dtype='complex128')# complex128
np.array([1, 4, 2, 5, 3], dtype='float64')    # float64
np.array([1, 4, 2, 5, 3], dtype='int64')      # int64
np.array([1, 4, 2, 5, 3], dtype='bool')       # bool 0=False, остальные=True

np.array([1, -4, 2, 0, 1j])                   # complex128

2. Что будет выведено?
np.array(..., dtype='str')[1] + '1'              # '41' строки соединяются
np.array(..., dtype='bool')[3] or array[1]       # -4  0=False или -4
np.array(..., dtype='bool')[3] and array[1]      # 0  0=False и вернет первое False
np.array([..., True], dtype='int64')[-1]         # 1  True превращается в 1
np.array([..., 1j], dtype='int64')[1]            # ошибка. комплексное нельзя перевести в int
np.array([..., 1j], dtype='str')[-1]             # '1j' строка
np.array([..., '3'])[-1] * 3                     # '333' умножение строки
