```python
import sys
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris

# Задача 1

def f(x):
    return x**2 - 2*x - 3

x1 = np.linspace(-10, 0, 100)
x2 = np.linspace(0, 7, 100)

plt.figure(figsize=(7, 5))
plt.plot(x1, f(x1), color='red', linestyle='-.', label='x ∈ [-10,0]')
plt.plot(x2, f(x2), color=(1.0,0.2,0.3), linestyle='-', label='x ∈ [0,7]')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Задача 1: Два графика функции')
plt.legend()
plt.grid(True)
plt.show()

# Задача 2

titanic = pd.read_csv("./data/titanic.csv")

# удаляем пропуски
df = titanic[['Age', 'Fare', 'Survived']].dropna()

plt.figure(figsize=(7, 5))

# выжившие
surv = df[df['Survived'] == 1]
plt.scatter(surv['Age'], surv['Fare'], marker='o', color='green', label='Выжил')

# умершие
nosurv = df[df['Survived'] == 0]
plt.scatter(nosurv['Age'], nosurv['Fare'], marker='x', color='red', label='Не выжил')

plt.xlabel('Возраст (Age)')
plt.ylabel('Цена билета (Fare)')
plt.title("Задача 2: Диаграмма рассеяния Titanic")
plt.legend()
plt.grid(True)
plt.show()

# Задача 3

iris = load_iris()
data = iris.data
target = iris.target
names = iris.feature_names

plt.figure(figsize=(12, 12))
for i in range(4):
    for j in range(4):
        plt.subplot(4, 4, i*4 + j + 1)
        plt.scatter(data[:, j], data[:, i], c=target, cmap='viridis')
        if i == 3:
            plt.xlabel(names[j])
        if j == 0:
            plt.ylabel(names[i])

plt.suptitle("Задача 3: Матрица диаграмм рассеяния Iris", fontsize=16)
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()

# Задача 4

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Гистограмма возраста
age = titanic['Age'].dropna()
axs[0].hist(age, bins=range(0, 90, 5))
axs[0].set_title("Распределение Age")
axs[0].set_xlabel("Возраст")
axs[0].set_ylabel("Количество")

# Гистограмма Fare для женщин
fare_female = titanic[titanic['Sex'] == 'female']['Fare'].dropna()
axs[1].hist(fare_female, bins=20)
axs[1].set_title("Распределение Fare (женщины)")
axs[1].set_xlabel("Fare")
axs[1].set_ylabel("Количество")

plt.suptitle("Задача 4: Гистограммы Titanic")
plt.tight_layout(rect=[0,0,1,0.94])
plt.show()

# Задача 5

df5 = titanic[['Pclass', 'Fare']].dropna().groupby('Pclass').mean()

plt.figure(figsize=(7, 5))
plt.bar(df5.index, df5['Fare'], label='Средняя стоимость Fare')
plt.xlabel('Класс каюты (Pclass)')
plt.ylabel('Средний Fare')
plt.title("Задача 5: Средняя стоимость билета по классам")
plt.legend()
plt.grid(True)
plt.show()

# Задача 6

df6 = titanic[['Pclass', 'Sex']].dropna()
count_male = df6[df6['Sex'] == 'male'].groupby('Pclass').size()
count_female = df6[df6['Sex'] == 'female'].groupby('Pclass').size()

plt.figure(figsize=(7, 5))
plt.bar(count_male.index, count_male, label='Мужчины')
plt.bar(count_female.index, count_female, bottom=count_male, label='Женщины')

plt.xlabel('Pclass')
plt.ylabel('Количество пассажиров')
plt.title("Задача 6: Распределение пассажиров по полу и классам")
plt.legend()
plt.grid(True)
plt.show()

# Задача 7

iris = load_iris()
target_counts = np.bincount(iris.target)
explode = [0, 0, 0]
explode[np.argmax(target_counts)] = 0.1

plt.figure(figsize=(6, 6))
plt.pie(target_counts, labels=iris.target_names, autopct='%1.1f%%', explode=explode)
plt.title("Задача 7: Круговая диаграмма Iris")
plt.legend(iris.target_names)
plt.show()

# Задача 8

plt.figure(figsize=(7, 5))
sns.histplot(titanic['Age'].dropna(), bins=20)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Задача 8: Гистограмма возраста Titanic (Seaborn)")
plt.show()
```

1. Для чего нужна диаграмма рассеяния?

Scatter plot применяется для:

Отображения взаимосвязи между двумя количественными переменными

Оценки степени корреляции

Поиска кластеров и выбросов

Изучения распределения точек в двумерном пространстве

Пример: определение зависимости веса от роста, анализ связи цены и качества.

2. Что такое гистограмма и чем она отличается от столбчатого графика?

Гистограмма — инструмент для отображения распределения непрерывной переменной:

X — интервалы (бины)

Y — частота значений в каждом интервале

Столбцы прилегают друг к другу (данные непрерывные)

Столбчатая диаграмма — используется для сравнения категориальных данных:

X — категории

Y — величины

Между столбцами есть промежутки (данные дискретные)

Основные отличия:

Свойство	          Гистограмма	    Столбчатая диаграмма
Тип данных	        Непрерывные	    Категориальные
Столбцы	            Соприкасаются	  Разделены
Порядок категорий	  Естественный	  Произвольный

4. Чем отличаются plt.Figure и plt.Axes?

Figure — это общий холст, содержащий один или несколько графиков.
Axes — отдельная область построения внутри Figure, где располагается конкретный график.

Аналогия: Figure — это лист бумаги, Axes — рисунок на нём.

fig = plt.figure()
ax = fig.add_subplot(111)

Пример:
fig, ax = plt.subplots()

fig.set_size_inches(10, 8)
fig.suptitle('Главный заголовок')

ax.plot(x, y)
ax.set_title('Название графика')
ax.set_xlabel('X')

10. На одном холсте нельзя разместить больше 16 графиков — так ли это?

Нет, это неверно.

Почему:

Количество субграфиков не ограничено 16.

fig, axes = plt.subplots(4, 5)

Размер холста можно регулировать.

plt.figure(figsize=(20, 16))

Отступы и масштаб тоже поддаются настройке.

plt.subplots_adjust(wspace=0.4, hspace=0.4)
