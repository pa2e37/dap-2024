Задача №1. https://leetcode.com/problems/two-sum/


```python
print("Введи числа через пробел")
nums = list(map(int, input().split()))

print("Введи таргет")
target = int(input())

def checker(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        diff = target - num
        if diff in seen:
            return seen[diff], i
        seen[num] = i

print(checker(nums, target))
```

Задача №2. https://leetcode.com/problems/palindrome-number/description/

```python
x = int(input("Введи число: "))

def solution(x):
    if x <= 0:
        return False

    original = x
    rev = 0

    while x:
        rev = rev * 10 + x % 10
        x //= 10

    return rev == original

print(solution(x))
```

Задача №3. https://leetcode.com/problems/roman-to-integer/description/

```python
roman_counts = {
    'I': 1,
    'V': 5,
    'X': 10,
    'L': 50,
    'C': 100,
    'D': 500,
    'M': 1000
}

def roman_to_int(s: str) -> int:
    total = 0
    prev_value = 0

    for char in reversed(s):
        value = roman_counts[char]
        if value < prev_value:
            total -= value
        else:
            total += value
        prev_value = value

    return total

s = input("Введи число римскими цифрами капсом: ")
print("Десятичное значение:", roman_to_int(s))
```

Задача №4. https://leetcode.com/problems/longest-common-prefix/description/

```python
strs = input("Введи слова через пробел: ").split()

def solution(strs):
    if not strs:
        return ""

    shortest = min(strs, key=len)
    for i, char in enumerate(shortest):
        if any(s[i] != char for s in strs):
            return shortest[:i]
    return shortest

print(solution(strs))
```

Задача №5. https://leetcode.com/problems/merge-two-sorted-lists/description/

```python
list1 = list(map(int, input("Список 1, вводи цифры через пробел: ").split()))
list2 = list(map(int, input("Список 2, вводи цифры через пробел: ").split()))

def merge_sorted(list1, list2):
    i = j = 0
    merged = []

    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1

    # Добавляем оставшиеся элементы
    merged.extend(list1[i:])
    merged.extend(list2[j:])
    return merged

print(merge_sorted(list1, list2))
```

Задача №6. https://leetcode.com/problems/longest-substring-without-repeating-characters/description/

```python
def solution(s: str) -> int:
    char_set = set()
    left = max_len = 0

    for right, char in enumerate(s):
        while char in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(char)
        max_len = max(max_len, right - left + 1)

    return max_len

s = input("Введи строку: ")
print("Длина самой длинной подстроки без повторов:", solution(s))
```

8. Как объявить приватный метод? Можно ли его вызвать за пределами класса?

Приватные методы в Python начинаются с двойного подчёркивания:

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
    
    def __validate_amount(self, amount):
        return amount > 0
    
    def deposit(self, amount):
        if self.__validate_amount(amount):
            self.__balance += amount


Вызывать можно, но это считается плохой практикой:

9. Как определить статический метод? Для чего он нужен?

Статический метод не связан ни с экземпляром, ни с классом — просто функция, помещённая внутрь класса.

class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def multiply(a, b):
        return a * b

result = MathUtils.add(5, 3)
