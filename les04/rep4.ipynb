```python
import pandas as pd
import numpy as np

def load_data(file_path):
    print("\n--- Задание 1: Чтение данных ---")
    df = pd.read_csv(file_path, index_col="PassengerId")
    print("Файл успешно загружен, индекс установлен на 'PassengerId'")
    return df

def head_data(df):
    print("\n--- Задание 2: Первые 6 записей ---")
    print(df.head(6))

def describe_data(df):
    print("\n--- Задание 3: Статистическое описание ---")
    print(df.describe(include='all'))

def gender_count(df):
    print("\n--- Задание 4: Количество мужчин и женщин ---")
    print(df['Sex'].value_counts())

def pclass_distribution(df):
    print("\n--- Задание 4: Распределение по классам Pclass ---")
    print("Распределение по классам (в целом):")
    print(df['Pclass'].value_counts())
    print("\nСтатистика классов по полу:")
    print(df.groupby('Sex')['Pclass'].value_counts())
    male_2nd_class = df[(df['Sex'] == 'male') & (df['Pclass'] == 2)].shape[0]
    print("\nКоличество мужчин во 2-м классе:", male_2nd_class)

def fare_stats(df):
    print("\n--- Задание 5: Медиана и стандартное отклонение стоимости билетов ---")
    median_fare = round(df['Fare'].median(), 2)
    std_fare = round(df['Fare'].std(), 2)
    print("Медианная стоимость Fare:", median_fare)
    print("Стандартное отклонение Fare:", std_fare)

def survival_by_age(df):
    print("\n--- Задание 6: Выживаемость по возрастным группам ---")
    young = df[df['Age'] < 30]
    old = df[df['Age'] > 60]
    young_surv = young['Survived'].mean()
    old_surv = old['Survived'].mean()
    print("Доля выживших младше 30 лет:", round(young_surv, 2))
    print("Доля выживших старше 60 лет:", round(old_surv, 2))

def survival_by_gender(df):
    print("\n--- Задание 7: Выживаемость мужчин и женщин ---")
    male_surv = df[df['Sex'] == 'male']['Survived'].mean()
    female_surv = df[df['Sex'] == 'female']['Survived'].mean()
    print("Процент выживших мужчин:", round(male_surv, 2))
    print("Процент выживших женщин:", round(female_surv, 2))

def most_popular_male_name(df):
    print("\n--- Задание 8: Наиболее распространённое мужское имя ---")
    male_names = df[df['Sex'] == 'male']['Name']
    first_names = male_names.apply(lambda x: x.split(',')[1].split('.')[1].strip())
    popular_name = first_names.value_counts().idxmax()
    print("Самое частое мужское имя:", popular_name)

def avg_age_by_class_gender(df):
    print("\n--- Задание 10: Средний возраст по полу и классу ---")
    print(df.groupby(['Pclass', 'Sex'])['Age'].mean().round(2))

def avg_age_survived(df):
    print("\n--- Задание 10: Средний возраст выживших и погибших ---")
    print(df.groupby('Survived')['Age'].mean().round(2))
    print("\nМожно заметить, что пассажиры 1-го класса старше, чем 2-го, а те — старше пассажиров 3-го класса")

file_path = "data/titanic.csv"
df = load_data(file_path)

head_data(df)
describe_data(df)
gender_count(df)
pclass_distribution(df)
fare_stats(df)
survival_by_age(df)
survival_by_gender(df)
most_popular_male_name(df)
avg_age_by_class_gender(df)
avg_age_survived(df)
```

1. Что такое медиана? Найти медиану для данного набора чисел.

Медиана — это значение, разделяющее упорядоченный набор данных на две равные части: половина элементов меньше медианы, половина - больше.

Если количество элементов нечётное: медианой является центральный элемент
Если чётное: медиана - это среднее двух центральных значений.

Пример: ряд [3, 1, 4, 2, 5]

Сортировка: [1, 2, 3, 4, 5]

Центральный элемент = 3

Медиана = 3

2. Что такое дисперсия? Найти дисперсию для предложенного ряда чисел.

Дисперсия - это среднее значение квадратов отклонений элементов ряда от его среднего.
Она характеризует, насколько сильно данные разбросаны вокруг среднего.

Пример: ряд [3, 1, 4, 2, 5]

Среднее: (3+1+4+2+5)/5 = 3

Квадраты отклонений:

(3−3)^2 = 0

(1−3)^2 = 4

(4−3)^2 = 1

(2−3)^2 = 1

(5−3)^2 = 4

Сумма: 10

Дисперсия = 10/5 = 2

Дисперсия = 2

3. Что такое среднеквадратичное отклонение? Найти СКО для набора чисел.

Среднеквадратичное отклонение - это квадратный корень дисперсии.
Оно показывает, насколько в среднем значения отклоняются от среднего.

Пример:
Дисперсия = 2
СКО = sqrt(2) = 1.414

СКО = 1.414

4. Что такое Series и чем он отличается от DataFrame?

Series — одномерная структура данных с индексами, хранящая элементы одного столбца.

DataFrame - двумерная таблица, состоящая из нескольких Series.

Главные различия:

Размерность: Series - 1D, DataFrame - 2D

Структура: Series содержит один столбец, DataFrame множество

Индексация: Series имеет единственный индекс, DataFrame индексы строк + названия столбцов

7. Что такое DataFrame и его отличие от Series?

DataFrame - табличная структура из строк и столбцов, где каждый столбец — отдельный Series.

Различия с Series:

DataFrame - 2D, Series - 1D

В DataFrame есть имена столбцов и индекс строк

Столбцы могут иметь разные типы данных

Операции могут применяться по строкам и столбцам
