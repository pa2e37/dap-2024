1. Типы ячеек в Jupyter Notebook

В Jupyter Notebook используются три основные разновидности ячеек:

Code — предназначены для запуска программного кода.

Markdown — позволяют оформлять текст с разметкой.

Raw — содержат необработанный текст без каких-либо преобразований.

Подробности о категориальных данных

Категориальные переменные принимают ограниченный набор значений.

Особенности:

конечное количество категорий

значения обозначают группы

могут быть упорядоченными или нет

Типы:

Номинальные (без порядка):

пол

цвет

город

Порядковые (имеют порядок):

степень образования

качество (плохо → отлично)

размер одежды

Основные способы обработки:

One-Hot Encoding — для категорий без порядка

Label Encoding — для упорядоченных значений

Frequency Encoding

8. Что такое Jupyter Notebook и для чего он нужен?

Jupyter Notebook — интерактивная среда, позволяющая объединять код, текст, визуализации и пояснения в одном документе.

Применяется для:

EDA

обучения и демонстраций

научных расчётов

быстрого прототипирования

Преимущества:

выполнение кода по частям

поддержка разных языков

удобная визуализация

возможность комбинировать текст и код
